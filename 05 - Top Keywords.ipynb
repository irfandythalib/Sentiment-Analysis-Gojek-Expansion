```python
import csv
import nltk
import re
from nltk import ngrams
from textblob import TextBlob
from nltk import collections
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

from nltk.corpus import stopwords
cachedStopWords = stopwords.words("english")

def concatenate_list_data(list):
    result= ''
    for element in list:
        result += str(element)
    return result

def extract_ngrams(data, num_of_grams):
    n_grams = TextBlob(data).ngrams(num_of_grams)
    return [ ' '.join(grams) for grams in n_grams]

```

## Top Keywords Positive

```python
words = []
target = 'dataset-machine-learning.csv'
with open(target,'r',encoding='utf-8') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    next(csvReader)
    
    
    for row in csvReader:
        sentence = row[4].lower()
        sentence = re.sub(r"http\S+", "", sentence)
        sentence = re.sub(r"pic.twitter.com\S+", "",sentence)
        sentence = re.sub(r"[^a-zA-Z0-9]+", ' ', sentence)
        sentence = re.sub(r"gojek", "",sentence)
        sentence = re.sub(r"grab", "",sentence)
        sentence = re.sub(r"singapore", "",sentence)
        sentence = re.sub(r"thailand", "",sentence)
        sentence = re.sub(r"indonesia", "",sentence)
        sentence = re.sub(r"malaysia", "",sentence)
        sentence = re.sub(r"vietnam", "",sentence)
        sentence = ' '.join([word for word in sentence.split() if word not in cachedStopWords])
#         print(sentence)
#         print(re.sub(r"[^a-zA-Z0-9]+", ' ', sentence))
        if row[2] == 'Positive':
            words.append(sentence)
```

```python
wordss = concatenate_list_data(words)
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(wordss)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
cleannegativetweetnew2 = []
for xxxx in words:
    cleannegativetweetnew2.extend(extract_ngrams(xxxx, 2))

esBigramFreq = collections.Counter(cleannegativetweetnew2)
most_common_exp=esBigramFreq.most_common(15)

for xasd in most_common_exp:
    print(xasd)
```
