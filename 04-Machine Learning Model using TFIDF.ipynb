```python
import pandas as pd
import numpy as np
import csv
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

#### Collect dat into Document and Classes List

```python
target = 'dataset-machine-learning.csv'
with open(target,'r',encoding='utf-8') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    text = []
    classes = []
    
    for row in csvReader:
        sentence = row[4].lower()
        sentenceclass = row[3]
        text.append(sentence)
        classes.append(sentenceclass)
```

#### Conventional Vectorizer

```python
data_vectorizer = CountVectorizer()
data_feature = data_vectorizer.fit_transform(text)
dataframe = pd.DataFrame(data = data_feature.todense(),columns = data_vectorizer.get_feature_names())
```

#### TFIDF Vectorizer

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_feature = tfidf_vectorizer.fit_transform(text)
dataframe2 = pd.DataFrame(data = tfidf_feature.todense(),columns = tfidf_vectorizer.get_feature_names())
```

#### Pandas Matrix to List

```python
document = dataframe2.values.tolist()
```

#### Machine Learning

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix

a_train, a_test, b_train, b_test = train_test_split(document, classes, test_size=0.33, random_state=42)
```

```python
# Neural Network model
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(13,13,13),max_iter=500)
model.fit(a_train,b_train)

predictions = model.predict(a_test)
print(classification_report(b_test,predictions))

scores = cross_val_score(model, document, classes, cv=5, scoring='accuracy')
print("Result : ", scores)
print("Mean : ", scores.mean())
print("Max : ", scores.max())
```

```python
# Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(a_train,b_train)

predictions = model.predict(a_test)
print(classification_report(b_test,predictions))

scores = cross_val_score(model, document, classes, cv=5, scoring='accuracy')
print("Result : ", scores)
print("Mean : ", scores.mean())
print("Max : ", scores.max())
```

```python
#Decision Tree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(a_train,b_train)

predictions = model.predict(a_test)
print(classification_report(b_test,predictions))

scores = cross_val_score(model, document, classes, cv=5, scoring='accuracy')
print("Result : ", scores)
print("Mean : ", scores.mean())
print("Max : ", scores.max())
```

```python
#SVM model
from sklearn import svm
model = svm.SVC(kernel='linear')
model.fit(a_train,b_train)

predictions = model.predict(a_test)
print(classification_report(b_test,predictions))

scores = cross_val_score(model, document, classes, cv=5, scoring='accuracy')
print("Result : ", scores)
print("Mean : ", scores.mean())
print("Max : ", scores.max())
```
